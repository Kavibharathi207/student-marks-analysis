```python
import pandas as pd
```

```python
df=pd.read_csv('Student_Marks.csv')
```

```python
print("Initial Datas: ")
print(df.head())
```

```python
top_math=df.loc[df['number_courses'].idxmax()]
print("Top Scores:")
print(top_math)
```

```python
df['Average']=df[['number_courses','time_study','Marks']].mean(axis=1)
print(df)
```

```python
df['Total']=df[['number_courses','time_study','Marks']].sum(axis=1)
print(df)
```

```python
df.to_csv('students_updated.csv',index=False)
print("\nUpdated DataFrame:")
print(df)
```

```python
df=pd.read_csv('students_updated.csv')
print(df)
```

```python
print(df.tail(7))
print("\nDataFrame after reading from CSV:")
```

```python
group_a=df[df['number_courses']>=8]
group_b=df[df['number_courses']<8]
print('Group A:')
print(group_a)
```

```python
print('Group B:')
print(group_b)
```

```python
group_a.to_csv('group_a.csv',index=False)
group_b.to_csv('group_b.csv',index=False)
print('Group A:')
print(group_a)
print('Group B:')
print(group_b)
```

```python
group_a_loaded=pd.read_csv('group_a.csv')
group_b_loaded=pd.read_csv('group_b.csv')
```

```python
merged_df=pd.concat([group_a_loaded,group_b_loaded],ignore_index=True)
```

```python
print(merged_df)
```

```python
merged_df.to_csv('merged_students.csv',index=False)
print("\nMerged DataFrames:")
print(merged_df)
```
